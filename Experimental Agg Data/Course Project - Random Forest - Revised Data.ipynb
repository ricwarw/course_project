```python
#import magic and libraries
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#charity data
charity = pd.read_csv('Charity3.csv')
charity['avg coll age'] = charity['avg coll age'].convert_objects(convert_numeric=True)
charity = charity.dropna(subset=['avg coll age'])
charity['postcode'] = charity['postcode'].astype(str)
charity.head()
```

```python
#ato data
tax = pd.read_csv('2014tax2.csv')
tax['postcode'] = tax['postcode'].astype(str)
charity_new = pd.merge(charity, tax, on='postcode')
charity_new.head()
```

```python
#abs data
burs = pd.read_csv('abs2011combo.csv')
burs['postcode'] = burs['postcode'].astype(str)
charity_plus = pd.merge(charity_new, burs, on='postcode')
charity_plus.head()
```

```python
#have a look at the data
charity_plus.describe(include='all')
```

```python
#Try Decision Trees
from sklearn import tree
from sklearn.cross_validation import train_test_split
from sklearn import metrics
```

```python
#split data up for training
X = charity_plus.filter(regex="[^postcode]")
y = charity_plus.postcode
#y_train = np.asarray(charity_new['total_received'], dtype="|S6")
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, dtype="|S6")
```

```python
# Create a decision tree classifier instance (start out with a small tree for interpretability)
ctree = tree.DecisionTreeClassifier(random_state=1, max_depth=2)
```

```python
# Fit the decision tree classifier
#y_train = np.asarray(charity_new['total_received'], dtype="|S6")
ctree.fit(X_train, y_train)
```
